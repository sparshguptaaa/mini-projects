```python
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------
# 1. Simulate coin tosses
# -------------------------------

np.random.seed(42)

num_tosses = 100
tosses = np.random.choice([0, 1], size=num_tosses)  # 1 = Head, 0 = Tail

# -------------------------------
# 2. Compute running probability
# -------------------------------

running_heads = np.cumsum(tosses)
running_probability = running_heads / np.arange(1, num_tosses + 1)

# -------------------------------
# 3. Plot convergence
# -------------------------------

plt.figure(figsize=(8, 5))
plt.plot(running_probability, label="Estimated P(Head)")
plt.axhline(y=0.5, color='r', linestyle='--', label="True P(Head) = 0.5")
plt.xlabel("Number of Tosses")
plt.ylabel("Probability of Head")
plt.title("Law of Large Numbers: Coin Toss Experiment")
plt.legend()
plt.grid(True)
plt.show()
```
